```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.xlabel('X Label')
plt.title('Hello World')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
#Object Oriented Method
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.9,.9])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes ([.2,.2,.8,.8])
axes2 = fig.add_axes ([.3,.6,.4,.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes ([0,0,1,1])
ax.plot(x,x**3,label='X Squared')
ax.plot(x,x**2,label='Y Squared')
ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'r--',x*2,x*3,'bs')
```
